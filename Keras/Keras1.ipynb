# Preprocess Data

```python
train_labels = list()
train_samples = list()
```

```python
# 0 shows the individual has no side effect, 1 side effect

for i in range(50):
    random_younger =randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
# keras expects train_set and train_labels in np.array format

train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
# keras expects scaled 
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1)) # for fit_transform of scaler we should use .reshape(-1,1)
```

```python
for i in scaled_train_samples:
    print(i)
```

### Preprocess is done

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
# model = Sequential([]) # expects an array which are layers, such as layer1, layer2, layer3 etc. This the first way
# we can also add layers such as model.add(layer1) etc. This is the second way
# model.add()
```

```python
model = Sequential([
    Dense(16, input_shape=(1,), activation='relu'), # first HIDDEN layer, (16) units: Positive integer, dimensionality of the output space.# shape is 1D
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
```

With the Keras Sequential model, you do not explicitly add an input layer. Instead, the first layer that you see in the video is actually the first hidden layer. Keras implicitly creates the input layer behind the scenes based on the input_shape parameter that you pass to the first hidden layer.

The choice of 16 and 32 neurons here was pretty arbitrary. Generally speaking, the more complex your data, the more layers and more neurons in each layer you'll likely need. Also, generally, the number of neurons increases with each layer as they layers become deeper in the network. There is not a general rule of thumb that I'm aware of that you can follow for choosing how many neurons to include in each layer. It's more of mixing trial and error along with experience from what's worked in previous models.

input_shape=(1,) means that the shape of each of my input samples is (1,). This is not the batch size.

32 is the number of nodes in this layer. This layer is actually the first hidden layer in the model, not the input layer. The input layer is implicitly created based on the input_shape parameter. This parameter tells Keras the shape of the data being passed to the model.﻿

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(scaled_train_samples, train_labels, batch_size=10, epochs=200, shuffle=True, verbose=2)
```

## Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = confusion_matrix(test_labels, rounded_predictions)
```

```python
# https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html

def plot_confusion_matix(cm, classes, 
                        normalize=False,
                        title='Confusion_matrix',
                        cmap = plt.cm.Blues):
    
    sns.set_context('poster', font_scale=0.8)
    plt.figure(figsize=(8,8))
    
    plt.imshow(cm, interpolation= 'nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks= np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
cm_plot_labels=['no_side_effects', 'had_side_effects']
plot_confusion_matix(cm, cm_plot_labels, title='Confusion_Matrix')
```
